```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data1).fit()
```

```python
model1.summary()
```

```python
#Observations 
# y cap = Beta-0, Beta-1
# y cap = Beta0+Beta1*x
# y cap = 13.8356 + 1.3397 * error
```

```python
# The probability(p-value) for (beta_o) is 0.707 > 0.05
# Therefore the intercept coefficient may not be that much dignificant in prediction
# However the p_value for "daily" (beta_1) is 0.00 < 0.05 
# Therfore the beta_1 coefficients is highly significient and is contribution to prediction.
```

```python
sns.regplot(x="daily", y="sunday",data=data1)
plt.xlim([0,1250])
plt.show()
```

```python
# predict in all given training data
pred = model1.predict(data1["daily"])
pred
```

```python
#Add predicted values as a column in data1
data1["Y_hat"] = pred
data1
```

```python
# Compute the error values residuals and add as another column
data1["residuals"] = data1["sunday"]-data1["Y_hat"]
data1
```
